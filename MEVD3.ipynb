**Датасет Supermarket sales** содержит информацию о продажах компании-супермаркета в 3 различных филиалах за 3 месяца. Этот датасет может быть полезен для решения ряда задач, связанных с анализом продаж и поведения покупателей. Например:

* Прогнозирование продаж: на основе данных о продажах в прошлом можно попытаться предсказать будущие продажи, прибыль, оптимизировать запасы товаров на складе.

* Анализ поведения покупателей: анализируя данные о продажах, можно выявить паттерны поведения покупателей и определить, какие категории товаров чаще всего покупаются, зависимость кол-ва покупок от типа покупателя, доволен или не доволен покупатель (по рейтингу) и т.д.

*Однако*, датасет Supermarket sales может быть не подходящим для некоторых задач, например:

* Анализ предпочтений потребителей: хотя данные о продажах могут дать представление о том, какие категории товаров пользуются спросом, они не позволяют получить информацию о том, почему покупатели выбирают именно эти товары. В датасете нет конкретных наименований, также, нельзя спрогнозировать, какие категории чаще всего покупают те или иные возрастные группы по причине отсутствия возраста в наших данных.

* Анализ конкурентной ситуации: датасет содержит только данные о продажах в одной компании-супермаркете, поэтому он может быть не очень полезен для анализа конкурентной ситуации на рынке в целом.

* Анализ факторов, влияющих на цены: хотя датасет содержит информацию о ценах на товары, он не содержит данных о факторах, влияющих на ценообразование, таких как затраты на производство или конкуренция на рынке.

# **Гипотеза 5. Утром делается меньше покупок, чем днем и вечером**

```python
df_sm['time'] = pd.to_datetime(df_sm['time'])

# Создание новых колонок 'day_part'
# с помощью метода apply() и функции, которая возвращает часть дня
def get_day_part(timestamp):
    hour = timestamp.hour
    if hour >= 0 and hour < 12:
        return 'morning'
    elif hour >= 12 and hour < 18:
        return 'day'
    elif hour >= 18 and hour < 24:
        return 'evening'


df_sm['day_part'] = df_sm['time'].apply(get_day_part)
```

```python
df_sm.groupby('day_part').agg({'quantity':'sum'})
```

```python
import altair as alt

source = df_sm.groupby('day_part').agg({'quantity':'sum'}).reset_index()

chart = alt.Chart(source).mark_bar().encode(
    x=alt.X('day_part:O', axis=alt.Axis(title='Day Part')),
    y=alt.Y('quantity:Q', axis=alt.Axis(title='Total Quantity')),
    color=alt.Color('day_part:O', legend=None)
).properties(
    title='Total Quantity by Day Part',
    width=alt.Step(40)  # Ширина столбцов
)

chart.configure_axis(
    labelFontSize=14,
    titleFontSize=16
).configure_title(
    fontSize=20
)
```

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
import pandas as pd

# Создаем датафрейм с данными
data = df_sm.groupby('product_line').agg({'unit_price':'median'}).reset_index()

# Найти индекс строки с самой низкой ценой
min_index = data['unit_price'].idxmin()

# Создаем столбец 'color' в датафрейме
data['color'] = 'navy'
data.loc[min_index, 'color'] = 'firebrick'

# Создаем источник данных
source = ColumnDataSource(data)

# Создаем фигуру
p = figure(x_range=data['product_line'], plot_height=400, plot_width=600, title='Median Unit Price by Product Line')
p.xaxis.axis_label = 'Product Line'
p.yaxis.axis_label = 'Median Unit Price'
p.xaxis.major_label_orientation = 0.4363

# Добавляем точки на график
p.circle(x='product_line', y='unit_price', size=12, source=source, color='color', alpha=0.8)

# Добавляем всплывающие подсказки
hover = HoverTool()
hover.tooltips = [('Product Line', '@product_line'), ('Median Unit Price', '@unit_price{0.2f}')]
p.add_tools(hover)

# Отображаем график
output_notebook()
show(p)
```
