```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('/content/bestsellers with categories.csv')
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ','_')
```

# **Top 10 best-selling books**

```python
df_top_10 = df.groupby('name')\
.agg({'author':'count'})\
.reset_index()\
.rename(columns={'author':'cnt'})\
.sort_values(by='cnt', ascending=False).head(10)
```

```python
df_top_10
```

# **Top 10 Authors**

```python
df_top_10a = df.groupby('author')\
.agg({'name':'count'})\
.reset_index()\
.rename(columns={'name':'cnt'})\
.sort_values(by='cnt', ascending=False).head(10)
```

```python
df_top_10a
```

```python
fig = px.bar(df_top_10a, x='author', y='cnt', title='Top 10 Authors', width=800, height=400)
fig.update_layout(xaxis_tickangle=-70)
fig.update_layout(width=800, height=600)
fig.show()
```

# **Rating**

```python
fig = px.histogram(df, x="user_rating",labels={'':'The Number of Books'},title="User Rating Histogram")
fig.show()
```

```python
def popularity(x):
  if x >= 4.8:
    return 'Extremely Popular'
  elif x >= 4.5:
    return 'Very Popular'
  elif x >= 4.0:
    return 'Fairly Popular'
  else:
    return 'Popular'
```

```python
df['popularity'] = df['user_rating'].apply(popularity)
```

```python
df.groupby('popularity').agg({'genre':'count'})
```

```python
grouped = df.groupby('popularity').agg({'genre':'count'}).reset_index()

fig = px.bar(grouped, x='popularity', y='genre', title='Distribution of Genres by Popularity',
             labels={'popularity':'Popularity', 'genre':'Number of Genres'})
fig.show()
```
