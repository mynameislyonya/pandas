```python
import pandas as pd  
```

```python
df = pd.read_csv('/content/test_data.csv') #считываю данные 
```

```python
df.info() #смотрю на тип данных
```

```python
df.head() #просто смотрю на датафрейм
```

предобработка данных

```python
df['Spend'] = df['Spend'].str.replace(",", ".") #заменяю запятую на точку
df['Spend'] = df['Spend'].astype(float).round(2) #преобразовывую в подходящий тип данных

df['Revenue_ad1d'] = df['Revenue_ad1d'].str.replace(",", ".")
df['Revenue_ad1d'] = df['Revenue_ad1d'].astype(float).round(2)

df['Revenue_in_app1d'] = df['Revenue_in_app1d'].str.replace(",", ".")
df['Revenue_in_app1d'] = df['Revenue_in_app1d'].astype(float).round(2)

df['Revenue_ad7d'] = df['Revenue_ad7d'].str.replace(",", ".")
df['Revenue_ad7d'] = df['Revenue_ad7d'].astype(float).round(2)

df['Revenue_in_app7d'] = df['Revenue_in_app7d'].str.replace(",", ".")
df['Revenue_in_app7d'] = df['Revenue_in_app7d'].astype(float).round(2)

df['Revenue_ad30d'] = df['Revenue_ad30d'].str.replace(",", ".")
df['Revenue_ad30d'] = df['Revenue_ad30d'].astype(float).round(2)

df['Revenue_in_app30d'] = df['Revenue_in_app30d'].str.replace(",", ".")
df['Revenue_in_app30d'] = df['Revenue_in_app30d'].astype(float).round(2)
```

## **Кампании в каких гео показали себя хуже всего?**

посчитаем конверсию в каждом регионе и найдем слабый

```python
df_geo = df.groupby('Geo', as_index=False).agg({'Impressions':'sum','Registrations':'sum'})
```

```python
df_geo['con'] = df_geo['Registrations']/df_geo['Impressions']*100
```

```python
df_geo.sort_values('con', ascending = True)
```

по конверсии от показа за регистрации можно увидеть, что australia oceania показал себя хуже всех.

# **ARPU**

Если это регистрации за 30 дней то можно посчитать ARPU

```python
df_arpu = df.groupby('Buyer', as_index=False).agg({'Revenue_in_app30d':'sum','Revenue_ad30d':'sum','Registrations':'sum'})
```

```python
df_arpu['arpu'] = (df_arpu['Revenue_in_app30d']+df_arpu['Revenue_ad30d'])/df_arpu['Registrations']
```

```python
df_arpu[['Buyer','arpu']]
```
